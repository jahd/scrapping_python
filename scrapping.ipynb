```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url="https://www.boulanger.com/c/refrigerateur?numPage=0"
response = requests.get(url)
```

```python
html = response.content
```

```python
soup = bs(html, "lxml")
```

```python
articles = soup.find_all("article", class_="col-24 product-item")
```

```python
article_informations = []
for article in articles:
    article_list = []
    name = article.find_all("h2", class_="product-item__label")[0].text
    price = article.find_all("p", class_="price__amount")[0].string
    keypoints = article.find_all("ul", class_="keypoints")[0].contents
    
    price = price.replace("\n", " ").replace(" ", "")
    name = name.replace(" ", "").replace("\n", " ")
    
    article_list.append(name)
    article_list.append(price)
    
    for keypoint in keypoints:
        my_content = keypoint.text.replace("\n", "").strip()
        if len(my_content) > 0:
            article_list.append(my_content)
            
    article_informations.append(article_list)
```

```python
df = pd.DataFrame(article_informations, columns = ['product_name', 'price', 'dimensions', 'decibels_energy','volume','size',"+"])
```

```python
# with first 10 pages : 
article_informations=[]

for i in range(1,10):
    url=f'https://www.boulanger.com/c/refrigerateur?numPage={i}'
    response = requests.get(url)
    html = response.content
    soup = bs(html, "lxml")
    
    articles = soup.find_all("article", class_="col-24 product-item")
    
    for article in articles:
        article_list = []
        name = article.find_all("h2", class_="product-item__label")[0].text
        price = article.find_all("p", class_="price__amount")[0].string
        keypoints = article.find_all("ul", class_="keypoints")[0].contents

        price = price.replace("\n", " ").replace(" ", "")
        name = name.replace(" ", "").replace("\n", " ")

        article_list.append(name)
        article_list.append(price)

        for keypoint in keypoints:
            my_content = keypoint.text.replace("\n", "").strip()
            if len(my_content) > 0:
                article_list.append(my_content)
        article_informations.append(article_list)
df = pd.DataFrame(article_informations, columns = ['product_name', 'price', 'dimensions', 'decibels_energy','volume','size',"+"])
```
